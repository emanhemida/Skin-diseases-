```python
# # import the necessary libraries
import os

# initialize an empty list to store the number of images in each class
num = []

# assuming 'labels' is a list containing class labels
for label in labels:
    # create the path to the folder for the current class
    path = train_path.format(label)
    
    # list all the files (images) in the folder
    folder_data = os.listdir(path)
    
    # initialize a counter to keep track of the number of images
    k = 0
    
    # print the label in uppercase for clarity
    print('\n', label.upper())
    
    # iterate through each image in the folder
    for image_path in folder_data:
        # increment the counter for each image
        k = k + 1
    
    # append the count (k) to the 'num' list
    num.append(k)
    
    # print the number of images in the current class
    print('there are ', k, ' images in ', label, 'class')
```

#Data Preprocessing

```python
#read img from training folder 
x_data =[]
y_data = []
for label in labels:
    path = train_path.format(label)
    folder_data = os.listdir(path)
    for image_path in folder_data:
        image = cv2.imread(path+image_path,cv2.IMREAD_COLOR)
        image_resized = cv2.resize(image, (64,64))
        x_data.append(np.array(image_resized))
        y_data.append(label)
```

#Data Augmentation using Albumentations

```python
# use albumentations
import albumentations as A
transform = A.Compose([
            A.Rotate(limit=40),
            A.RandomBrightness(limit=0.1),
            A.ImageCompression(quality_lower=85, quality_upper=100, p=0.5),
            A.HueSaturationValue(hue_shift_limit=20, sat_shift_limit=30, val_shift_limit=20, p=0.5),
            A.RandomContrast(limit=0.2, p=0.5),
            A.HorizontalFlip(),
        ])

for label in labels:
    path = train_path.format(label)
    folder_data = os.listdir(path)
    for image_path in folder_data:
        image = cv2.imread(path+image_path,cv2.IMREAD_COLOR)
        # Augment an image
        transformed = transform(image=image)
        transformed_image = transformed["image"]
        image_resized = cv2.resize(transformed_image, (64,64))
        x_data.append(np.array(image_resized))
        y_data.append(label)
```

```python
num = []
for label in labels:
    k = 0
    print('\n', label.upper())
    for image_path in x_data:
        k = k+1
    num.append(k)
    print('there are ', k,' images in ', label, 'class')
```

```python
plt.figure(figsize = (5,5))
plt.bar(labels, num)
plt.title('NUMBER OF IMAGES CONTAINED IN EACH CLASS')
plt.xlabel('classes')
plt.xticks(rotation=90)
plt.ylabel('count')
plt.show()
```

```python
fig,ax=plt.subplots(5,2)
fig.set_size_inches(13,13)
for i in range(5):
    for j in range (2):
        l=rn.randint(0,len(y_data))
        ax[i,j].imshow(x_data[l])
        ax[i,j].set_title('Disease is: '+y_data[l])
        
plt.tight_layout()
```

```python
x_data = np.array(x_data)

y_data = np.array(y_data)


print('shape of X is: ', x_data.shape, ' Y is: ', y_data.shape)
```

```python
le=LabelEncoder()
Y=le.fit_transform(y_data)
Y=to_categorical(Y,num_cls)
x_data = x_data/255   #standarization 
Y.shape
```

```python
from tensorflow.keras.applications import ResNet50
conv_base = ResNet50(weights='imagenet', 
                        include_top=False, 
                        input_shape=(64, 64, 3))
```

#Model Building

```python
#build model  
model = Sequential()

model.add(Conv2D(filters = 16, kernel_size = (5,5),padding = 'Same',activation ='relu', input_shape = (64,64,3)))

model.add(Conv2D(filters = 32, kernel_size = (3,3),padding = 'Same',activation ='relu'))

model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',activation ='relu'))
# model.add(conv_base)
model.add(Flatten())
model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_cls, activation = "softmax"))
```

```python
# Making the conv_base NOT trainable:

for layer in conv_base.layers[:]:
    layer.trainable = False
```

#Model Compilation

```python
from tensorflow.keras.optimizers import SGD

model.compile(optimizer = SGD(lr=1e-3),
              loss='categorical_crossentropy',metrics=['acc'])
```

```python
model.summary()
```

#Data Preparation for Testing

```python
#read img from test folder 
test_label = os.listdir('test')
print(test_label)
```

```python
x_test =[]
y_test = []

for label in test_label:
    path = test_path.format(label)
    folder_data = os.listdir(path)
    for image_path in folder_data:
        image = cv2.imread(path+image_path,cv2.IMREAD_COLOR)
        image_resized = cv2.resize(image, (64,64))
        x_test.append(np.array(image_resized))
        y_test.append(label)
```

```python
x_test = np.array(x_test)
x_test = x_test/255
y_test = np.array(y_test)
le = LabelEncoder()
y = le.fit_transform(y_test)
y = to_categorical(y,num_cls)
```

```python
x_test.shape,y.shape
```

#Training the Model

```python
# Setting up early stopping
from keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor='val_loss', patience=7, verbose=1, mode='auto')
```

```python
History = model.fit(x_data,Y,epochs = 100, validation_data = (x_test,y),
                    verbose = 1,callbacks=[early_stop])
```

 Data augmentation during training using ImageDataGenerator

```python
datagen = ImageDataGenerator(
        rotation_range=90,
        featurewise_center=True,
        featurewise_std_normalization=True,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

datagen.fit(x_data)

val_datagen = ImageDataGenerator(
        rotation_range=90,
        featurewise_center=True,
        featurewise_std_normalization=True,
         width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

val_datagen.fit(x_test)
```

```python
History = model.fit(datagen.flow(x_data,Y, batch_size=32),
                              epochs = 100, 
                              validation_data = val_datagen.flow(x_test,y,batch_size=32),
                              verbose = 1, 
                              steps_per_epoch=x_data.shape[0] // 32,
                              callbacks=[early_stop])
```

#Model Evaluation and Visualization

```python
plt.plot(History.history['loss'])
plt.plot(History.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.plot(History.history['acc'])
plt.plot(History.history['val_acc'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```
